# Question 4

- Time complexity of this code is O(n), as it performs all the logic in 1 single for loop

- Space complexity of this code is O(2n), which is O(n), as we are using dict(), we store duplicate records

- set() doesn't provide us to store duplicate records, hence I chose dict() for my logic


```python
#import necessary packages
import collections
from timeit import timeit

#define the get_indices function
def get_indices(data):
    
    #create a defaultdict to store the results
    results = collections.defaultdict(list)
    read = dict()
    #iterate over the input list of tuples 
    for i, uid in enumerate(data):
        
        #check for input format, if incorrect format is passed, say int or string instead of tuple throw error
        if type(uid) != tuple:
            
            print('Please provide a proper tuple input format. e.g. [(123,),(''asc'',)]')
            break

        #iterate individual tuple to fetch tuple elements
        for detail in uid:
            
            #store the index position for the user
            #try and except is written for exception handling
            try:
                idx = read[detail]
                break
            except KeyError:
                pass
        
        #append index value to the list if there are duplicate values
        else:
            idx = len(results)
        results[idx].append(i)
        
        #add the index position if there is no duplicate record (considered as unique value)
        for detail in uid:
            if detail not in read:
                read[detail] = idx

    return results.values()
    return [results[key] for key in sorted(results.keys())]

#code run starts here
if __name__ == "__main__":
    
    #test results using different inputs
    data = [('id1', 'addr1', 'pw1'), ('idx', 'addr1', 'pwx'), ('idz', 'addrz', 'pwz'),('idy', 'addry', 'pwx')]
    print(get_indices(data))
    data = [('id1', 'addr1', 'pw1'), ('idx', 'addr1', 'pwx'), ('idz', 'addrz', 'pwz',123),('idy', 'addry', 'pwx'),('id3',123)]
    print(get_indices(data))
    data = [(123,),(),(123,),(234,)]
    print(get_indices(data))
    data = [(123,),(),(123,),(234,),('fgdfgdh')]
    print(get_indices(data))
    data = [(123,),(),(123,),(234,),('fgdfgdh',)]
    print(get_indices(data))
    
    #verify the time complexity of the function
    print('Time taken for execution using defaultdict is', timeit(lambda: get_indices(data), number=1000))
```
